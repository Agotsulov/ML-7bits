```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn import model_selection
from sklearn import neighbors
from sklearn import ensemble
from sklearn import tree
from sklearn import svm 
from sklearn import metrics
%matplotlib inline

```

```python
df = pd.DataFrame(np.random.randint(low=0, high=2, size=(100, 2)))
df.head()
   
```

```python
df_and = df.copy()

a = []
for i in df_and.index:
    a.append(df_and.loc[i][0] and df_and.loc[i][1])
    
df_and['and'] = a
df_and.head()
```

```python
X = df
y = df_and['and']
```

```python
MO_models = [neighbors.KNeighborsClassifier(n_neighbors=3), \
             ensemble.RandomForestClassifier(n_estimators=10,n_jobs=5), \
             linear_model.SGDClassifier(), \
             tree.DecisionTreeClassifier(max_depth=2)]
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.35)
```

```python
for m in MO_models:
    m.fit(X_train, y_train)
    y_pred = m.predict(X_test)
    print(m)
    print(metrics.classification_report(y_test, y_pred))
```

```python
def convert_base(num, to_base=10, from_base=10):
    # first convert to decimal number
    if isinstance(num, str):
        n = int(num, from_base)
    else:
        n = int(num)
    # now convert decimal to 'to_base' base
    alphabet = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    if n < to_base:
        return alphabet[n]
    else:
        return convert_base(n // to_base, to_base) + alphabet[n % to_base]
```
