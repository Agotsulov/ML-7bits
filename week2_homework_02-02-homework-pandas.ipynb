12) Посчитайте среднее время работы (hours-per-week) зарабатывающих мало и много (salary) для каждой страны (native-country).

```python
import numpy as np

df_less = df[df['income'] == "<=50K"] 
df_more = df[df['income'] == ">50K"]

countries = df['native.country'].drop_duplicates().dropna().values

mean_hpw_less = [0 for i in range(0, np.size(countries))]
mean_hpw_more = [0 for i in range(0, np.size(countries))]

for i in range(0, np.size(countres)):
    mean_hpw_less[i] = (df_less[df_less['native.country'] == countries[i]])['hours.per.week'].mean()
    mean_hpw_more[i] = (df_more[df_more['native.country'] == countries[i]])['hours.per.week'].mean()
    
print("Меньше 50К")
for i in range(0, np.size(countres)):
    print(countres[i], end=' ')
    print(mean_hpw_less[i])

print(" ")
print("Больше 50К")
for i in range(0, np.size(countres)):
    print(countres[i], end=' ')
    print(mean_hpw_more[i])
```
